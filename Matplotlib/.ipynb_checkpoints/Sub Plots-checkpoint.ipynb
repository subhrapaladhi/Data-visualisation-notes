# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)

y=x**2
```

```python
# fig, axes = plt.subplots(nrows=1, ncols=2)
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
fig
```

```python
axes
```

```python
for ax in axes:
    for bx in ax:
        bx.plot(x, y, 'b')
        bx.set_xlabel('x')
        bx.set_ylabel('y')
        bx.set_title('title')

fig
```
