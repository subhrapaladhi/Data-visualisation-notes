# Plot appearance

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# lw = linewidth
# ls = linestyle

ax.plot(x, y, color='orange', lw=3, alpha=0.3, ls='--', drawstyle="steps")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,marker='o', markersize=10,markerfacecolor='red',markeredgewidth=2,markeredgecolor='yellow')
```

## Control over axis appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)

ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
