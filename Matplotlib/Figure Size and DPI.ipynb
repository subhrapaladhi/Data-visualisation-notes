# Figure Size and DPI

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure(figsize = (4,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5,4))

axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
```

## Saving a Figure

```python
fig.savefig('my_picture.png', dpi=200)
```
