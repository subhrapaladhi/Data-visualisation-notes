```python
import seaborn as sns
import numpy as np

%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot=True)
```

```python
fp = flights.pivot_table(index='month', columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,linecolor='white',linewidth=0.1)
```

```python
sns.clustermap(fp)
```
