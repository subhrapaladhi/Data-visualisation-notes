```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Histograms

Histograms show the distribution of observations among various ranges. 

```python
sns.distplot(tips['total_bill'])
```

The following histograms show that most of the bills lie in the range 10-20

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

## Joint plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

### hexagons

Darker colour of hexagons means more values in that area

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

### Regression

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

### KDE

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

## Pair plot

```python
sns.pairplot(tips,hue='sex')
```

## Rugplot

```python
sns.rugplot(tips['total_bill'])
```
